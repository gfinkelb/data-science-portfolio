Question 1. The half life of Cesium 90, a dangerous radioactive isotope released in the Fukushima reactor accident, is 30 years. That is, half of its radioactive material will decay every 30 years. Rearrange the following lines to produce a program that computes when 99 percent of the material is gone.

```python
#Question 1 code
year = 0
percentage = 100
while percentage > 1.0 :
    year = year + 30
    percentage = percentage * 0.5
print (year)
```

Question 2. Write a while loop that prints all positive numbers that are divisible by 10 and less than a given number n. For example, if n is 100, print 10 20 30 40 50 60 70 80 90, one value per line.

```python
#Question 2 code
n = int(input("n: "))
current_number = 10
while current_number < n:
    print(current_number)
    current_number += 10
```

Question 4. Write a program that prints all digits of any integer in reverse order, one per line.


```python
#Question 4 code
def print_digits_in_reverse(number):
    # Convert the number to a string
    num_str = str(number)

    # Iterate through the characters in reverse order
    for digit in reversed(num_str):
        # Print each digit on a new line
        print(digit)

# Get user input for an integer
user_input = input("Enter an integer: ")

try:
    # Convert user input to an integer
    user_integer = int(user_input)

    # Call the function to print digits in reverse order
    print_digits_in_reverse(user_integer)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```
