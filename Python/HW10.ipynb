Question 3. Complete the following program that removes all negative numbers from a list of values.

```python
#Question 3 code
values = [17, 25, -5, 30, 100, 96, -48, 5, 14, -30]
i = 0
while i < len(values):
    if values[i] < 0:
        del values[i]
    else:
        i += 1
print(values)
```
