Question 4. Complete this program that prints the position of the last uppercase letter in a string. Print -1 if the word has no uppercase letters.




```python
#Question 4 code

word = input("word: ")
position = len(word) - 1  
found = False

while position >= 0:
    ch = word[position]
    if ch.isupper():
        found = True
        break
    position -= 1

if found:
    print("Position:", position)
else:
    print("Position: -1")
```

Question 5. Walk through the following code that counts how many adjacent duplicates are in a sequence of non-negative integer inputs, using a negative value as the sentinel. Assume the inputs are 1 2 2 4 4 4 3 -1. 



```python
#Question 5 code

duplicates = 0
value = int(input("Enter a value: "))
while value >= 0 :   
    previous = value   
    value = int(input("Enter a value: "))   
    if value == previous :      
        duplicates = duplicates + 1
print(duplicates)
```
