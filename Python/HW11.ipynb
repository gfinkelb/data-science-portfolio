Question 1. Complete the following function that determines whether the set passed to the function is a subset of the set of rainbow colors.



```python
#Question 1 code
def main():
    colors = ["red", "orange", "blue", "green"]
    if inRainbow(colors):
        print("rainbow colors")
    else:
        print("not rainbow colors")

def inRainbow(colors):
    rainbow = {"red", "orange", "yellow", "green", "blue", "indigo", "violet"}
    return set(colors).issubset(rainbow)

main()
```

Question 2.  Given three baskets of fruit represented as Python sets, complete the following program using set operations to determine which fruits are in basket1 or basket2 but not in basket3.



```python
#Question 2 code
basket1 = {"apples", "oranges", "grapes"}
basket2 = {"blueberries", "pears", "apples", "oranges", "lemons"}
basket3 = {"bananas", "lemons", "grapes", "apples"}

fruit = (basket1 | basket2) - basket3

for element in sorted(fruit):
    print(element, end=", ")

print()
```

Question 3. Your task is to write a program that reads a text string from the user and determines which letters of the alphabet are not used in the string.




```python
#Question 3 code
inputStr = input("Enter a text string: ")
text = inputStr.upper()

used_letters = set()

for char in text:
    if char.isalpha():  
        used_letters.add(char)

alphabet = set("ABCDEFGHIJKLMNOPQRSTUVWXYZ")

notUsed = alphabet - used_letters

for element in sorted(notUsed):
    print(element, end="")

print()
```

Question 5. Complete the following program that acts as a translator for the periodic table of the elements. The program repeatedly prompts the user for an element symbol and then prints the elements name or an error message until the user enters Q.

Your program should use the prompt Enter symbol (or Q to quit): for user input. The output should be Invalid symbol if the user enters an invalid element symbol; otherwise the output is the translation in the format H = Hydrogen.




```python
#Question 5 code
# The periodic table of elements.
periodicTable = {
   "H": "Hydrogen", "C": "Carbon", "O": "Oxygen",
   "Fe": "Iron", "Au": "Gold", "Na": "Sodium",
   "Ag": "Silver", "Pb": "Lead", "Pu": "Plutonimum"
}

# Use the prompt "Enter symbol (or Q to quit): "
# Print either "Invalid symbol." if an error occurs
# or "element = element name" ("%s = %s")

while True:
    element = input("Enter symbol (or Q to quit): ")
    if element.upper() == "Q":
        break  # Exit loop if user enters "Q" to quit
    
    if element in periodicTable:
        print(f"{element} = {periodicTable[element]}")
    else:
        print("Invalid symbol.")
```
